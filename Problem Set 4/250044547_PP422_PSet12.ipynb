# sklearn Pipeline Function

While presented in some prior code, a useful `sklearn` function for PCA and later analysis will be the `Pipeline`. Pipeline allows you to create a sequence of data instructions that can be applied to your data before fitting a model. It is particularly useful for ensuring your training and testing data both receive the same series of transformations. 

For example, when fitting a PCA model, we first need to standardize our data, then estimate and store the principal components ($Z_1, Z_2,\dots, Z_m$). If we subsequently wish to use these principal components to make estimates on testing data, we need to ensure the testing data undergoes the same scaling process and that the same principal component loading factors are applied to transform the columns of the testing data. Creating a pipeline is an efficient way to control the application of these steps.

Read the pipeline [user guide](https://scikit-learn.org/stable/modules/compose.html#pipeline) and examine the following two examples related to PCA:

1. [Selecting dimensionality reduction](https://scikit-learn.org/stable/auto_examples/compose/plot_compare_reduction.html#sphx-glr-auto-examples-compose-plot-compare-reduction-py)

## Acemoglu and Johnson - Rebalancing AI

Read the provided piece by Acemoglu and Johnson. This piece draws on the authors' larger book _Power and Progress_ which examines historical waves of technological progress. A big theme of this work is understanding to what extent, if any, is the current rise of AI different from previous technological innovations, and what the historical impacts of these technological revolutions have been.

After reading the article, explain what the authors mean when they discuss the importance of 'new tasks' and how this shapes the potential long-run impact of automation from AI. The authors also differentiate between productive automation and 'so-so automation.' What is this distinction? Are there any sectors beyond those referenced in the piece that you think are particularly susceptible to automation that fails to create new tasks? 



Acemoglu and Johnson argue that the long-run impact of automation depends critically on whether technological change creates new tasks for workers. If automation replaces labour without generating meaningful productivity gains or new forms of work, it can lead to higher inequality and a declining labour share of income.

The authors emphasize that technological progress historically created new tasks, which absorbed displaced workers and increased the marginal productivity of labour. When new tasks emerge, workers can move into these roles, wages can rise, and the benefits of technological progress are more broadly shared. For example, the rise of the automotive industry did not only replace horse-related occupations; it also created entirely new sectors and tasks, including car manufacturing, maintenance, infrastructure development, and even new leisure and recreation industries linked to mobility.

The authors distinguish between productive automation and “so-so automation.” Productive automation increases overall productivity and creates new complementary tasks for workers. It raises the marginal productivity of labour and generates new employment opportunities. In contrast, so-so automation replaces workers with machines without significantly increasing productivity or creating new tasks. An example of so-so automation is the introduction of self-checkout kiosks in grocery stores, where workers are replaced but overall productivity gains are limited.

There are many sectors which are suspectable to automation which would fail to create new tasks 
- Autonomous delivery drones can replace human couriers in last-mile logistics without necessarily generating significant new employment opportunities
- AI-powered customer service systems can automate complaint handling and routine inquiries, reducing the need for human representatives.
- The shift toward autonomous vehicles could replace drivers without necessarily creating enough new employment opportunities to offset job losses, particularly once the supporting infrastructure is established.
